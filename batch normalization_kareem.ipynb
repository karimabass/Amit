```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from tensorflow import keras

```

```python
dataset = pd.read_csv(r'E:\machine learning course\lec 34 DL 3 George\Session-3-\Churn_Modelling.csv')
dataset.head()
```

```python
dataset.shape
```

```python
dataset.describe()
```

```python

dataset.drop('Geography',axis=1,inplace=True)
dataset.drop('Gender',axis=1,inplace=True)
X = dataset.iloc[:, 3:11].values
y = dataset.iloc[:, 11].values
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
model = keras.models.Sequential([
    keras.layers.Dense(3, activation="tanh",input_dim = 8),
    keras.layers.BatchNormalization(),
    keras.layers.Dense(5, activation="tanh"),
    keras.layers.BatchNormalization(),
    keras.layers.Dense(6, activation="tanh"),
    keras.layers.BatchNormalization(),
    keras.layers.Dense(4, activation="tanh"),
    keras.layers.BatchNormalization(),
    keras.layers.Dense(1, activation="sigmoid")
])
```

```python
model.summary()
```

```python
model.compile(optimizer = 'Adam', loss = 'binary_crossentropy', metrics = ['accuracy','recall'])
```

```python
model.fit(X_train, y_train, batch_size = 32, epochs = 100)
```
